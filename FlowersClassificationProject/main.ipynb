```python
import os
import cv2
x = []
y = []
for folder in os.listdir('flowers'):
  for file in os.listdir(os.path.join('flowers', folder)):
    if file.endswith('jpg'):
      y.append(folder)
      img = cv2.imread(os.path.join('flowers', folder, file))
      img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      im2 = cv2.resize(img_rgb, (128, 128))
      x.append(im2)    
```

```python
import numpy as np
data_arr = np.array(x)
label_arr = np.array(y)
```

```python
!pip install sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
encoder = LabelEncoder()
y = encoder.fit_transform(label_arr)
y = to_categorical(y, 5)
x = data_arr / 255
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Activation, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam,SGD,Adagrad,Adadelta,RMSprop

SIZE = 128

model = Sequential()
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',activation ='relu', input_shape = (SIZE,SIZE,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = 'Same',activation ='relu'))
model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = 'Same',activation ='relu'))
model.add(Conv2D(filters = 128, kernel_size = (3,3),padding = 'Same',activation ='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(5, activation = "softmax"))
```

```python
datagen = ImageDataGenerator(
        rotation_range=20,
        zoom_range = 0.20,
        width_shift_range=0.3,
        height_shift_range=0.3,
        horizontal_flip=True,
        vertical_flip=True)

datagen.fit(X_train)
```

```python
import requests
inception_net = tf.keras.applications.InceptionV3()
response = requests.get("https://git.io/JJkYN")
labels = response.text.split("\n")
def flower_classification(img):
    img = img.reshape((1, SIZE, SIZE, 3))
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    prediction = inception_net(img).flatten()
    return {labels[i]: float(prediction[i]) for i in range(1000)}

image = gr.inputs.Image(shape=(128, 128))
label = gr.outputs.Label(num_top_classes=3)

gr.Interface(fn=flower_classification, inputs=image, outputs=label, capture_session=True).launch()
```
